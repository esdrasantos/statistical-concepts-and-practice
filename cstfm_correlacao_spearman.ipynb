```python
import pandas as pd
import numpy as np
```

```python
vendas = pd.read_excel('/content/gdrive/MyDrive/Curso de Estatística 2.0/Arquivos Excel/Dados3_CursoEstatistica.xlsx', sheet_name=3)

vendas
```

```python
print('N:', len(vendas))
```

```python
from scipy.stats import spearmanr

corr, p = spearmanr(vendas['Preço'], vendas['Dias'])

print('Correlação de Spearman:', corr)
print('p-valor:', p)
```

# Interpretação
[Quadro de Correlações](https://www.researchgate.net/profile/Anelise-Figari/publication/307857769/figure/fig2/AS:668947959525381@1536501030240/Quadro-8-Interpretacao-das-escalas-de-coeficientes-de-correlacao.png)

```python
if abs(corr) == 1.0:
  print('Perfeita')
elif abs(corr) > 0.7 and abs(corr) < 1.0:
  print('Muito forte')
elif abs(corr) < 0.7 and abs(corr) > 0.5:
  print('Substancial')
elif abs(corr) < 0.5 and abs(corr) > 0.3:
  print('Moderada')
elif abs(corr) < 0.3 and abs(corr) > 0.1:
  print('Fraca')
elif abs(corr) < 0.1:
  print('Muito fraca')
elif abs(corr) == 0:
  print('Nenhuma')
```

```python
if corr > 0:
  print('Positiva')
elif corr < 0:
  print('Negativa')
```
