# Wicoxon & Mann-Whitney para 2 amostras *dependentes*
Testa se as mediana de duas amostras são estatisticamente iguais.
O exemplo compara par a par em função de Anos, por isso, dizemos que são dependentes.

```python
retorno = pd.read_excel('/content/drive/MyDrive/Curso de Estatística 2.0/Arquivos Excel/Dados3_CursoEstatistica.xlsx', sheet_name=4)

retorno
```

```python
from scipy.stats import wilcoxon

# Teste para 2 Amostras
# 2 Amostras bicaudal é o padrão desse teste

stat, p = wilcoxon(retorno['Metais'], retorno['Renda'])

print('stat=%.3f, p=%.3f' % (stat, p))
```

```python
# Interpretação
alfa = 0.05

if p > alfa:
  print('Não rejeita-se H0')
  print('Logo as medianas são iguais.')
else:
  print('Rejeita-se H0')
  print('Logo as medianas não são iguais')
```

# Wicoxon & Mann-Whitney para 2 amostras *independentes*:
Testa se as mediana de duas amostras são estatisticamente iguais.

```python
profissao = pd.read_excel('/content/drive/MyDrive/Curso de Estatística 2.0/Arquivos Excel/Dados3_CursoEstatistica.xlsx', sheet_name=5)

profissao
```

```python
from scipy.stats import mannwhitneyu

stat, p = mannwhitneyu(profissao['Finanças'], profissao['Computação'])

print('stat=%.3f, p=%.3f' % (stat, p))
```

```python
# Decisão
alfa = 0.05

if p > alfa:
  print('Não rejeita-se H0')
  print('Logo as medianas são iguais.')
else:
  print('Rejeita-se H0')
  print('Logo as medianas são diferentes.')
```
